```python
import matplotlib.pyplot as plt
import networkx as nx

# Beispielgraph als Dictionary anlegen (Adjazenzliste mit Gewichten)
graph = {
    'A': [('B', 2), ('C', 7)],
    'B': [('A', 2), ('C', 6), ('D', 1)],
    'C': [('A', 7), ('B', 6), ('D', 2), ('E', 8)],
    'D': [('B', 1), ('C', 2), ('E', 4)],
    'E': [('C', 8), ('D', 4)]
}
```

```python
# Beispielgraph visualisieren
G = nx.Graph()
for node, edges in graph.items():
    for neighbor, weight in edges:
        G.add_edge(node, neighbor, weight=weight)

pos = nx.spring_layout(G, seed=7)
labels = nx.get_edge_attributes(G, 'weight')

plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1200, font_size=14)
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Beispielgraph mit Gewichten")
plt.show()
```
